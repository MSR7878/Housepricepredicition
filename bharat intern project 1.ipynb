```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
sns.set()
import numpy as np

dataset = pd.read_csv("C:/Users/rath/Downloads/Housing.csv")
print(dataset.head(10))# this one is used fo printing first 10 entries of the dataset
```

```python
dataset.info()
```

```python
dataset.describe(include ='all')
#describe will tell about  statistics tools
#parameter include=all will display NaN values as well
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
cat = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
def binary_map(x):
    return x.map({"yes":1,"no":0})
dataset[cat] = dataset[cat].apply(binary_map)
```

```python
dataset.head()
```

```python
#now for dummy variable
table = pd.get_dummies(dataset['furnishingstatus'])
table.head()
```

```python
table = pd.get_dummies(dataset['furnishingstatus'], drop_first = True)  #recreate table but now drop the first column(furnished)
table.head()
```

```python
dataset = pd.concat([dataset,table], axis =1)
dataset.head()
```

```python
dataset['semi-furnished'] = dataset['semi-furnished'].map({True: 1, False: 0})
dataset['unfurnished'] = dataset['unfurnished'].map({True: 1, False: 0})
```

```python
dataset.head()
```

```python
dataset.drop(['furnishingstatus'], axis=1, inplace = True)
dataset.head()
```

```python
#spliting data into training and testing data
dataset.columns
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(0) 
df_train, df_test = train_test_split(dataset, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
df_train.head()
```

```python
#scaling traning dataset 
from sklearn.preprocessing import MinMaxScaler #transform the features of a dataset so that they fall within a specific range, typically between 0 and 1
scaler = MinMaxScaler()
```

```python
var_to_scale = ['area','bedrooms','bathrooms','stories','parking','price']
#for all numerical columns(not dummy)
```

```python
df_train[var_to_scale] = scaler.fit_transform(df_train[var_to_scale])
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
#Now training the model
y_train = df_train.pop('price')
x_train = df_train
```

```python
y_train.head()
```

```python
lm=LinearRegression()
lm.fit(x_train, y_train)
```

```python
lm.coef_
```

```python
lm.score(x_train,y_train)
```

```python
var_to_scale = ['area','bedrooms','bathrooms','stories','parking','price']
```

```python
df_test[var_to_scale] = scaler.fit_transform(df_test[var_to_scale])
```

```python
#NOW TESTING THE MODEL USING TEST DATA
y_test = df_test.pop('price')
x_test = df_test
```

```python
#prediction 
predictions = lm.predict(x_test)
```

```python
predictions
```

```python
#accuracy
from sklearn.metrics import r2_score
r2_score(y_test, predictions)
```
